# Compare Results

Load result JSON files, compute summary metrics, and inspect method-level comparison.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().resolve()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from scripts.evaluate_results import compare_results
from shared.io_utils import write_json

RESULTS_DIR = PROJECT_ROOT / "results"
summary = compare_results(RESULTS_DIR)
write_json(RESULTS_DIR / "summary_metrics.json", summary)
summary
```

import pandas as pd

df = pd.DataFrame(summary)
df[[
    "method",
    "total",

    "avg_latency_seconds",

```python
error_df = (
    pd.json_normalize(df["error_counts"])
    .set_index(df["method"])
    .stack()
    .rename("count")
    .reset_index()
    .rename(columns={"level_1": "error_type"})
)
error_df
```
